```python

from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
import matplotlib.pyplot as plt
import pandas as pd
import json
import math
import re
from typing import Any, Dict, List, Optional, Tuple, Literal

import numpy as np
from pydantic import BaseModel, ValidationError

from langchain_ollama import ChatOllama
```

```python
df = pd.read_csv(
    "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/ZNoKMJ9rssJn-QbJ49kOzA/student-mat.csv",
)
```

```python
df.head(5)
```

```python
df.info
```

```python
llm = ChatOllama(model="qwen2.5:7b-instruct", temperature=0)
```

```python
agent = create_pandas_dataframe_agent(
    llm,
    df,
    verbose=True,
    return_intermediate_steps = True,
    handle_parsing_errors = True,allow_dangerous_code=True  )
```

```python
response = agent.invoke("how many rows of data are in this file")
```

```python
response['output']
```

```python
response = agent.invoke("Generate a pie chart to display the average value of walc fo each gender")
```

```python
response = agent.invoke("Generate a bar chart to display number of men and women")
```

```python
response = agent.invoke("summarize the data")
```
